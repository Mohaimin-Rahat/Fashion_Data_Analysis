```python
df.groupby('product_type_name')['product_type_no'].count().reset_index()
```

```python
plt.subplots(figsize=(15,10))

ax=sns.countplot("product_group_name",data=df,palette='dark')
ax.set_xlabel(xlabel='iffhvgfldhvb',fontsize=15)
ax.set_ylabel(ylabel='iffhvgfldhvb',fontsize=15)
ax.set_title(label='iffhvgfldhvb',fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.subplots(figsize=(15,10))
sns.heatmap(df.corr(),annot=True)
```

```python
from wordcloud import WordCloud 
```

```python
wc=WordCloud(height=1080,width=1920,background_color='white')
```

```python
wc.generate(" ".join(df.product_group_name))
plt.subplots(figsize=(15,10))
plt.imshow(wc)
plt.axis("off")
```

```python
plt.figure(figsize=(20,25))
var=1
for i in df:
    if var <= 15:
        ax=plt.subplot(5,3,var)
        sns.distplot(df[i],bins=100)
        plt.xlabel(i,fontsize=20)
    var+=1
plt.show()
```
